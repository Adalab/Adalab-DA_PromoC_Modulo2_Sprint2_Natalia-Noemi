- Extraed los valores únicos su frecuencia. BONUS presenta los resultados en un dataframe. Ver un ejemplo del resultado en apuntes.

```python
def val_unicos(df):

    lista_unicos = []
    lista_columnas = []

    for col in df.columns:
        lista_columnas.append(col)
        lista_unicos.append(df[col].unique())

        dicc = {"columna": lista_columnas, "valores unicos": lista_unicos}

        df_unicos_categoricas = pd.DataFrame(dicc)

    return df_unicos_categoricas
```

```python
val_unicos(ataques)
```
