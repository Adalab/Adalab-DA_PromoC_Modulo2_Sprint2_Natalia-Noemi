```python
#Vemos cuantos nulos tenemos por columna
ataques.stb.missing()
```

```python
country_freq = ataques.stb.freq(["country"], thresh = 70)
country_freq
```

- Comprobamos que el 70% de los ataques se ha producido en USA y Australia, lo que nos puede hacer pensar que es en estos países donde hay más tiburones. 

```python
ataques.stb.freq(["country", 'sex'], thresh = 70)
```

- Observamos que el 70% de los ataques ha sido a hombres y en USA, Australia y Sudáfrica. Esto podría deberse a que los hombres practican más deportes acuáticos. 

```python
ataques.stb.freq(["activity", "fatal"], thresh = 60)
```

- En la relación entre las actividades y la mortalidad de los ataques vemos que dentro del 60% de estos, la mayoría se ha producido haciendo surf y/o nadando. Apreciamos además que nadar resulta más peligroso ya que la mitad de los ataques han tenido un final fatal.

```python
ataques.stb.freq(["mes", "fatal"], thresh = 60)
```

- Vemos que los meses con más ataques son Octubre y Agosto. Esto podría deberse a las temperaturas del agua. Podríamos hacer un anális más exhaustivo para comprobar esta hipótesis añadiendo también la columna country.
